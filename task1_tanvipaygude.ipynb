```python
import pandas as pd
```

```python
data=pd.read_csv("car_prices.csv")
```

```python
data
```

```python
print("Missing values before cleaning:\n", data.isnull().sum())
data=data.dropna()
```

```python
data
```

```python
# 2. Remove duplicate rows
data = data.drop_duplicates()
data
```

```python
# 3. Standardize text values (example for text columns)
text_columns = ['make', 'model', 'trim', 'body', 'transmission', 'state', 'color', 'interior', 'seller']
for col in text_columns:
    if col in data.columns:
        data[col] = data[col].astype(str).str.lower().str.strip()
```

```python
data
```

```python
# 4. Convert date formats to a consistent type (e.g., dd-mm-yyyy)
data['saledate'] = pd.to_datetime(data['saledate'], errors='coerce', utc=True)
data['saledate'] = data['saledate'].dt.tz_localize(None).dt.strftime('%d-%m-%Y')

```

```python
data
```

```python
# 5. Rename column headers to be clean and uniform
data.columns = [col.lower().replace(" ", "_") for col in data.columns]
data
```
